# 数据集说明（2026-03 更新）

当前流程默认使用 ISCX 数据，先组装为会话级样本，再进入字节会话模型。

关键路径：

- 原始数据：`costSensitive/dataset/iscx/`
- 会话样本：`costSensitive/processed_full/sessions/`
- 推理输出：`costSensitive/pytorch_model/session_predictions.csv`

补充：Unknown 聚类基于 `session_predictions.csv + session_embeddings.npy` 生成。

你的目标下，**最通用且最适配当前项目（原始 pcap → 字节序列建模）**的数据集组合是这三类：

**首选组合（推荐）**
- **ISCX VPN-nonVPN**：最适合你现在的“加密/非加密与应用识别”主线，通常有原始 pcap，能直接接你现有预处理脚本。
- **USTC-TFC2016**：包含正常+恶意流量（含加密场景），对“入侵/恶意通信检测”很实用，且常被用于字节级建模对比。
- **CIC-IDS2018（或 CIC-IDS2017）**：入侵检测基准数据，攻击类型覆盖广，适合验证“工程可部署性”和泛化能力。

**为什么这套最通用**
- 都是网络安全研究里常见基准，论文可比性好。  
- 以 **pcap/流量级** 为主，符合你“无需解密、端到端自动学习”的方法路线。  
- 能同时覆盖：正常流量、异常流量、恶意攻击流量三类目标。

**与你项目的匹配度**
- 你当前代码链路是“pcap→图像/IDX→训练推理”，所以优先选**能拿到原始 pcap**的数据集。  
- 这三类最容易做成统一标签体系（如 normal / vpn / attack 子类），便于你做多任务评估。

**建议的实验设计（针对“未知流量识别”）**
- 做 **Open-set**：训练只用已知类，测试加入未见类（未知应用/未知攻击）。  
- 做 **Cross-dataset**：例如在 ISCX+USTC 训练，在 CIC 测试，验证泛化。  
- 做 **时间切分**：按采集日期切分训练/测试，模拟真实上线漂移。
