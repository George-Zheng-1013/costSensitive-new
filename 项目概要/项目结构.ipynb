# 项目结构（精简视图）

```text
Encryption-traffic-detection/
├─ run_full_chain.py                # API + engine 一键启动
├─ README.md
├─ services/
│  ├─ api_server.py                 # FastAPI 服务
│  ├─ backend_engine.py             # 消费 realtime jsonl 写 SQLite
│  ├─ ai_analyst.py                 # LLM 分析告警与行为
│  └─ ai_config.py
├─ costSensitive/
│  ├─ session_data.py               # pcap -> 会话样本 manifest
│  ├─ train_byte_session.py         # 训练入口
│  ├─ inference.py                  # 离线推理入口
│  ├─ build_centroid_detector.py    # OOD 质心检测器构建
│  ├─ main_realtime.py              # 实时推理入口（live）
│  ├─ realtime/
│  │  ├─ model_def.py               # 模型定义
│  │  ├─ flow_assembler.py          # 实时会话组装
│  │  ├─ single_pipeline.py         # 实时推理 + 可解释性
│  │  ├─ unknown_detector.py        # OOD 距离检测
│  │  └─ config.py
│  ├─ processed_full/sessions/      # 构建后的会话数据
│  ├─ pytorch_model/                # 模型、评估与实时输出
│  └─ dataset/iscx/                 # 原始 pcap 数据
└─ web-app/
   ├─ src/views/                    # 大屏页面
   ├─ src/services/                 # API 与 websocket 客户端
   └─ src/stores/
```

说明：当前核心链路是“字节级会话建模”，旧图像化 28x28 训练链路已不作为主流程。
