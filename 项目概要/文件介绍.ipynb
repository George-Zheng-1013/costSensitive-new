# 文件介绍（当前主干）

## 训练与离线推理
- `costSensitive/session_data.py`：从 pcap/pcapng 组装会话样本，输出 manifest 和 npz 样本。
- `costSensitive/train_byte_session.py`：训练 `ByteSessionClassifier`，输出模型与训练日志。
- `costSensitive/inference.py`：离线推理，输出预测 CSV、embedding NPY、评估报告 JSON。
- `costSensitive/build_centroid_detector.py`：基于 train embedding 构建 OOD 质心阈值检测器。

## 实时推理
- `costSensitive/main_realtime.py`：实时推理入口（仅支持 `--mode live`）。
- `costSensitive/realtime/model_def.py`：`PacketEncoder + SessionEncoder + Classifier` 模型定义。
- `costSensitive/realtime/flow_assembler.py`：实时流组装，按包数/FIN/RST/超时触发会话样本。
- `costSensitive/realtime/single_pipeline.py`：推理主流水线，输出 CSV/JSONL，含异常分数与可解释性字段。
- `costSensitive/realtime/unknown_detector.py`：质心距离异常检测逻辑与构建工具。

## 后端服务
- `run_full_chain.py`：一键启动入口，负责设置环境变量并启动 API。
- `services/api_server.py`：FastAPI 接口层（概览、告警、XAI、AI insight、WebSocket）。
- `services/backend_engine.py`：消费实时 JSONL 并写入 SQLite (`netguard_logs.db`)。

## 前端
- `web-app/`：Vue + Vite 前端。
- `web-app/src/views/`：`OverviewView`、`AlertsView`、`XaiView`、`AiInsightsView` 等页面。

## 关键产物目录
- `costSensitive/processed_full/sessions/`：会话样本、manifest、label_map。
- `costSensitive/pytorch_model/`：模型权重、训练日志、推理报告、实时输出。
